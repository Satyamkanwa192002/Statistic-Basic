1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales?

Qualitative Data: Describes qualities or characteristics.
Nominal Scale: Categorizes data without any order (e.g., gender, types of fruits).
Ordinal Scale: Categorizes with a meaningful order (e.g., satisfaction levels like dissatisfied, neutral, satisfied).
Quantitative Data: Represents numerical values.
Interval Scale: Numerical with equal intervals, but no true zero (e.g., temperature in Celsius).
Ratio Scale: Numerical with a meaningful zero, allowing for meaningful comparisons (e.g., weight, height).

2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate?

Mean: The average, best used with symmetric data distributions (e.g., average age of participants).
Median: The middle value, useful with skewed data or outliers (e.g., median household income).
Mode: The most frequent value, helpful for categorical data (e.g., most common car color).

3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

Dispersion: Shows the spread of values around the central value.
Variance: Measures the average squared difference from the mean.
Standard Deviation: Square root of variance, indicating the average distance from the mean in the same units as the data.

4. What is a box plot, and what can it tell you about the distribution of data?

Box Plot: Visual representation of data distribution showing the median, quartiles, and outliers. It indicates spread, symmetry, and skewness by dividing data into quartiles and marking any outliers.

5. Discuss the role of random sampling in making inferences about populations.

Random Sampling: Ensures every member of a population has an equal chance of selection, helping to make unbiased inferences. It provides a representative subset, minimizing bias and supporting generalizations to the larger population.

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

Skewness: Indicates asymmetry in data distribution.
Positive Skew (Right): Tail on the right, mean > median (e.g., income distribution).
Negative Skew (Left): Tail on the left, mean < median (e.g., age at retirement).
Skewness affects interpretation as it suggests that data might not be normally distributed, impacting measures like mean.

7. What is the interquartile range (IQR), and how is it used to detect outliers?

Interquartile Range (IQR): Difference between the first (Q1) and third quartiles (Q3), showing the spread of the middle 50% of data. Values 1.5 * IQR above Q3 or below Q1 are considered outliers.

9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

Normal Distribution: Symmetrical, bell-shaped curve with mean, median, and mode equal.
Empirical Rule:
68% of data within 1 standard deviation of the mean.
95% within 2 standard deviations.
99.7% within 3 standard deviations.

10. Provide a real-life example of a Poisson process and calculate the probability for a specific event.

Poisson Process: Models events occurring independently at a constant rate, like the number of calls a call center receives per hour.
Example Calculation: If the average call rate is 5 calls per hour, we can calculate the probability of receiving exactly 7 calls in a given hour using the Poisson formula.

11. Explain what a random variable is and differentiate between discrete and continuous random variables.

Random Variable: A variable representing outcomes of a random event.
Discrete: Takes specific values, like the number of students in a class.
Continuous: Can take any value within a range, like height or temperature.

Example Dataset: Suppose we have height and weight data for a group of people.
Covariance: Indicates if height and weight increase or decrease together.
Correlation: Shows the strength and direction of this relationship, with values closer to 1 indicating a strong positive relationship.
